# [문제] 주민번호를 입력 받아 검증하는 프로그램을 작성하시오

­ 다음 페이지를 참고하여 사용자 정의 메소드를 확인하고 필요한 메소드를 작성할 것.
­ 주민번호는 “-“ 포함 14자리여야 한다.
­ 생년월일 뒤에 “-“이 포함되어 있어야 한다.
­ “-“ 앞과 뒤에는 숫자 리터럴만 가능하다.
­ 성별에 해당하는 8번째 위치는 1~4까지의 값만 가능 (1, 3:남자, 2, 4: 여자)
­ 검증 코드는 아래의 알고리즘으로 구현하여 확인
­ 문자열과 관련된 다양한 메소드를 이용하여 작업한다.
* 유효성 검사
- 최종 결과물 : 1991년생, 12월 25일, 남자

```python
# False 리턴 : 잘못된 주민번호
# True 리턴 : 올바른 주민번호 : 2000년 1월 1일 생 남자 22세 입니다.
# 주민번호 가운데 -
# 주민번호 길이 14자리
# - 앞,뒤에는 숫자만
# 성별 위치는 1, 2, 3, 4외에는 불가
# 주민번호를 공식에 넣어서 마지막 digit과 같아야만 올바른 주민번호
def jumin_validation(jumin):
    if len(jumin) != 14 or jumin[6] != '-' or not jumin.split('-')[0].isnumeric() or not jumin.split('-')[1].isnumeric():
        return False

    gender = jumin[7]
    if not(gender == "1" or gender == "2" or gender == "3" or gender == "4"):
        return False
    
    year = jumin[:2] 
    year = "19" + year if int(gender) == 1 or int(gender) == 2 else "20" + year
    month = jumin[2:4] 
    day = jumin[4:6] 
    genderType = "여자" if int(gender) % 2 == 0 else "남자"
    age = 2023 - int(year)
    
    temp = "%s년생 %s월 %s일 %d세 %s입니다." % (year, month, day, age, genderType)
    return temp

jumin = input("주민등록번호 : ")
result = jumin_validation(jumin)

if(not result):
    print("잘못된 주민번호입니다.")
else:
    print(result)
```
